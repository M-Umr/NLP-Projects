```python
import nltk
def file_handling(filename):
    file=open(filename,encoding="utf8")
    text=file.read()
    return text
def tokenize_file_return(file_for_tokenize):
    nltk_tokens=nltk.word_tokenize(file_for_tokenize)
    return nltk_tokens
def make_sentence(token):
    sentence=token[0]
    for word in token[1:len(token)]:
        sentence= sentence+" "+word
    return sentence
def bigrams(text):
    bi = []
    token_address = 0
    for token in text[:len(text) - 1]:
        bi.append(token +" " +text[token_address + 1])
        token_address += 1
    return bi

def trigrams(text):
    tri = []
    token_address = 0
    for token in text[:len(text) - 2]:
        tri.append(token +" " + text[token_address + 1] +" " + text[token_address + 2])
        token_address += 1
    return tri
def ngrams(tokenize_text,grams):  
    model=[]  
    count=0  
    for token in tokenize_text[:len(tokenize_text)-(grams-1)]: 
        model.append(make_sentence(tokenize_text[count:count+grams]))
        count=count+1  
    return model
def word_count(token):
    counts = dict()
    words = token

    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1

    return counts
def probability_repeat_word(count):
    
    for word in count:
        count[word]=count[word]/len(count)
    
    return count 
def set_order(token):

        ordered_tokens = set()
        result = []
        for word in token:
            if word not in ordered_tokens:
                ordered_tokens.add(word)
                result.append(word)
                #print(result)
        return result
def generate_sentence(word,count):
    
    large=0
    global predict
    for text in count:
        texts=text.split(" ")
        next_text=texts[0]
        if word == next_text and count[text]>large:
            
            predict =texts[1:len(texts)]
            large=count[text]  
    if predict is not None:
        return make_sentence(predict)
    else:   
        return "Not in Dictionary"
file_return=file_handling("robot_text.txt")
tokenize_file=tokenize_file_return(file_return)
x=input("1 for ngram\n 2 for bigram\n 3 for trigram\n")
if x==1:
    make_ngram=ngrams(tokenize_file,5)
if x==2:
    make_ngram=bigrams(tokenize_file,2)
if x==3:
    make_ngram=trigrams(tokenize_file,3)
# make_ngram=ngrams(tokenize_file,3)
word_counting=word_count(make_ngram)
find_probability=probability_repeat_word(word_counting)
next=input("Give word to make sentence  ")
text=next.split(" ")
ans=[]
ans.append(next+" ")    
for word in range(3):
    text=generate_sentence(text[-1],find_probability)
    ans.append(text+" ")
ans=set_order(ans)
for text in ans:
    print(text,end="")
```
